```python
import numpy as np
import csv
import torch
import torch.nn.functional as F
import pandas as pd
import sklearn as skl
import sklearn.svm as svm
```

```python
train=pd.read_csv('ascertain_multi_train.csv')
test=pd.read_csv('ascertain_multi_test.csv')
```

```python
test
```

```python
# Get training data in SVM form
trainlabelArousal = train.label_Arousal
trainlabelValence = train.label_Valence
trainsvmform = train.iloc[:,range(4,137)]
trainsvmform = trainsvmform.fillna(0)
trainsvmform.isnull().sum().sum()
```

```python
#Use SVM function
SVMfuncA = svm.SVC()
SVMfuncV = svm.SVC()
SVMfuncA.fit(trainsvmform,trainlabelArousal)
SVMfuncV.fit(trainsvmform,trainlabelValence)
```

```python
# Get testing data in SVM form
testlabelArousal = test.label_Arousal
testlabelValence = test.label_Valence
testsvmform = test.iloc[:,range(4,137)]
testsvmform = testsvmform.fillna(0)
testsvmform.isnull().sum().sum()
```

```python
#Predictions
predArousal = SVMfuncA.predict(testsvmform)
predValence = SVMfuncV.predict(testsvmform)
checksA = (predArousal == testlabelArousal)
checksV = (predValence == testlabelValence)
```

```python
print(checksA.sum()/checksA.size)
print(checksV.sum()/checksV.size)
```
